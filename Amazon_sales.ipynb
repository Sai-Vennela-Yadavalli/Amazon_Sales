```python
df.Date = pd.to_datetime(df["Date"], format="%d-%m-%y")
print(df)
```

```python
# To remove unwanted column
df.drop(columns=["Order ID"], inplace=True)
```

```python
df
```

```python
#Ensure Price and Total Sales are floats:
df['Price'] = df['Price'].astype(float)
df['Total Sales'] = df['Total Sales'].astype(float)
df['Quantity'] = df['Quantity'].astype(int)

```

```python
df.groupby('Date')['Total Sales'].sum().plot(figsize=(12,6))
plt.title('Total Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.show()
```

```python
# Prepared top 10 products
top_products = df.groupby('Product')['Total Sales'].sum().sort_values(ascending=False).head(10).reset_index()

# Setting style
sns.set_style("darkgrid")

# Ploted using Seaborn with dark gradient color and no warning
plt.figure(figsize=(10,6))
sns.barplot(data=top_products, x='Product', y='Total Sales', hue='Product', palette='Set2', legend=False)

# Customising
plt.title('Top 10 Bestselling Products', fontsize=16)
plt.xlabel('Product Name')
plt.ylabel('Total Sales')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
df.groupby('Category')['Total Sales'].sum().plot(kind='bar', color='green')
plt.title('Sales by Category')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
df['Payment Method'].value_counts().plot(kind='pie', autopct='%1.1f%%', figsize=(8,8))
plt.title('Payment Method Distribution')
plt.ylabel('')
plt.show()
```

```python
df['Status'].value_counts().plot(kind='bar', color='purple')
plt.title('Order Status Count')
plt.xlabel('Status')
plt.ylabel('Number of Orders')
plt.show()
```

```python
df['Customer Location'].value_counts().head(10).plot(kind='bar', color='red')
plt.title('Top 10 Customer Locations')
plt.xlabel('Location')
plt.ylabel('Number of Orders')
plt.show()
```
